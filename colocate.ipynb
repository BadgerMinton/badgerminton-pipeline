```python
import pandas as pd

df = pd.read_csv(
    "https://raw.githubusercontent.com/BadgerMinton/badgerminton-data/refs/heads/main/elo-scores/2024-11-07_elo_scores.csv"
)

# Filter based on min 10 games played
result_df = df[df["played"] >= 10].reset_index(drop=True)
result_df.index = pd.RangeIndex(start=1, stop=len(result_df) + 1)
result_df
```
