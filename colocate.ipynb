```python
import pandas as pd

df = pd.read_csv(
    "https://raw.githubusercontent.com/BadgerMinton/badgerminton-data/refs/heads/main/elo-scores/2025-01-23_elo_scores.csv"
)

result_df = df[df["played"] >= 20].reset_index(drop=True)
result_df.index = pd.RangeIndex(start=1, stop=len(result_df) + 1)

display(result_df)
# Filter based on min 20 games played
# 23 January 2025
```

```python
result_df.sort_values(by=["event_change"])
```
